# Using Selenium we will open the above link and then later using Beautiful Soup we can collect the Headings and Sub Headings from the data 

```python
import re
import requests
from bs4 import BeautifulSoup

url = 'https://www.careerguide.com/career-options'

html_text = requests.get(url).text
with open('name.html', 'w',encoding='utf-8') as file:
    file.write(html_text)
with open('name.html', 'r',encoding='utf-8') as html_file:
    content = html_file.read()
    #Taking content from the website using BeautifulSoup
    soup = BeautifulSoup(content, 'lxml')
    blocks = soup.find_all('div', class_='col-md-4')
    d={}
    for block in blocks:
        try:
            #Find heading
            heading = block.find('h2', class_='c-font-bold')
            #Find all subheadings
            subheadings = block.find_all('li')
            print('')
            print(f'Heading is: {heading.text}')
            print('')
            print(f'Subheadings are:')
            #Printing and storing headings in a dictionary d 
            for subheading in subheadings:
                try:
                    d[heading.text]+=[subheading.text]
                except:
                    d[heading.text]=[subheading.text]
                print(subheading.text)
                pass
        except:
            continue
```

# Once again using Beautiful Soup we'll collect the Role, Company Name and Location

```python
job = driver.find_element_by_xpath('//input[@aria-label="Search by title, skill, or company"]')
location_val = driver.find_element_by_xpath('//input[@aria-label="City, state, or zip code"]')
location_val.send_keys('Andhra Pradesh')
for i in range(len(jobs_list)):
    job.clear()
    job.send_keys(jobs_list[i])
    job.send_keys(Keys.ENTER)
    content = driver.page_source
    soup = BeautifulSoup(content, 'lxml')
    profiles = soup.findAll('div', class_='artdeco-entity-lockup__content')
    for profile in profiles:
        role = profile.find(
            'a', class_='disabled ember-view job-card-container__link job-card-list__title')
        company = profile.find('div', class_="artdeco-entity-lockup__subtitle")
        location = profile.find('div', class_="artdeco-entity-lockup__caption")
        role_value = role.text.strip()
        company_value = company.text.strip()
        l = location.text.split()
        location_value = ' '.join(l)
        print(f'Role: {role_value}')
        print(f'Company: {company_value}')
        print(f'Location: {location_value}')
        s=f'Role: {role_value}, Company: {company_value}, Location: {location_value}'
        try:
            data[jobs_list[i]]+=[s]
        except:
            data[jobs_list[i]]=[s]
        print("")
    
```
